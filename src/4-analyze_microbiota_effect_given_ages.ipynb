# Differentially expressed genes

Extract DEG between GF and SPF for the different ages
- Threshold for adjusted p-value: 0.05
- Threshold for adjusted significant fold change: 1.5

Table with the factors / constrasts

```r
sub_contrasts = contrasts %>%
    filter(Info == 'GF vs SPF (Young)' | 
           Info == 'GF vs SPF (Old)')
sub_contrasts
```

```r
deg_results = lapply(sub_contrasts$Info, function(x) get_dge_results(x, dge, sub_contrasts))
names(deg_results) = sub_contrasts$Info
save(deg_results, file=paste(dir_path, "deg_results.RData", sep=''))
```

Extract the log2FC of the DEG

```r
deg = extract_DEG_log2FC(deg_results, dir_path)
```

```r
all_deg_genes = unique(c(all_deg_genes, deg$sign_fc_deg$genes))
save(all_deg_genes, file=paste(dir_path, "all_deg_genes.RData", sep=''))
```

## Z-score

```r
comps = list(
    "GF vs SPF (Young)" = c(metadata %>% filter(Age == "Young" & Microbiota == "SPF") %>% pull(sample),
                  metadata %>% filter(Age == "Young" & Microbiota == "GF") %>% pull(sample)),
    "GF vs SPF (Old)" = c(metadata %>% filter(Age == "Old" & Microbiota == "SPF") %>% pull(sample),
                  metadata %>% filter(Age == "Old" & Microbiota == "GF") %>% pull(sample))
)
comps
```

### Column order: age  -  microbiota

```r
plot_z_score_heatmap(z_scores,
                     deg$sign_fc_deg$genes,
                     col_order$am,
                     annot_col$am,
                     "All DE genes in age comparison for any microbiota",
                     col_order$am)
pdf(paste(dir_path, "z_score_am.pdf", sep=''))
plot_z_score_heatmap(z_scores,
                     deg$sign_fc_deg$genes,
                     col_order$am,
                     annot_col$am,
                     "All DE genes in age comparison for any microbiota",
                     col_order$am)
dev.off()
```

```r
for(comp in names(comps)){
    plot_z_score_heatmap(z_scores,
        deg$sign_fc_deg %>% filter(!is.na((!!as.name(comp)))) %>% pull(genes),
        col_order$am,
        annot_col$am,
        paste("DE genes in", comp),
        comps[[comp]])
}
```

### Column order: microbiota - age

```r
plot_z_score_heatmap(z_scores,
                     deg$sign_fc_deg$genes,
                     col_order$ma,
                     annot_col$ma,
                     "All DE genes in age comparison for any microbiota",
                     col_order$ma)
pdf(paste(dir_path, "z_score_ma.pdf", sep=''))
plot_z_score_heatmap(z_scores,
                     deg$sign_fc_deg$genes,
                     col_order$ma,
                     annot_col$ma,
                     "All DE genes in age comparison for any microbiota",
                     col_order$ma)
dev.off()
```

```r
for(comp in names(comps)){
    plot_z_score_heatmap(z_scores,
        deg$sign_fc_deg %>% filter(!is.na((!!as.name(comp)))) %>% pull(genes),
        col_order$ma,
        annot_col$ma,
        paste("DE genes in", comp),
        comps[[comp]])
}
```
